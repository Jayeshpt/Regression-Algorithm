```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Salary_Data.csv')
df.head()
```

```python
x=df['YearsExperience']
y=df['Salary']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scalar_x=StandardScaler()
scalar_y=StandardScaler()
x=scalar_x.fit_transform(x.values.reshape(-1,1))
y=scalar_y.fit_transform(y.values.reshape(-1,1))
```

```python
plt.scatter(x,y,color = "blue")
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.svm import SVR
svr_reg=SVR()
svr_reg.fit(x,y)
```

```python
plt.scatter(x,y,color = "blue")
plt.plot(x,svr_reg.predict(x),color='red')
plt.title('Support Vector Regression')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
y_pred =svr_reg.predict([[7.9]])
print(y_pred)
```

```python
y_pred = y_pred.reshape(-1, 1)
y_pred = scalar_y.inverse_transform(y_pred)
y_pred
```
